```python
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Load Breast Cancer Wisconsin (Diagnostic) dataset
breast_cancer = pd.read_csv("breast cancer.csv")

# Create a DataFrame with features and target
data_input = breast_cancer.drop(['id', 'diagnosis'], axis=1)
data_target = breast_cancer['diagnosis']

data_input
```

```python
# Number of subsets
num_subsets = 10

# split training and testing
# fold 1
data_input_training_subset1 = data_input.drop(data_input.index[range(57)])

data_target_training_subset1 = data_target.drop(data_target.index[range(57)])

data_input_testing_subset1 = data_input[0:57]

data_target_testing_subset1 = data_target[0:57]


# fold 2
skiprowfoldF2 = [n for n in range(57,114)]
data_input_training_subset2 = data_input.drop(data_input.index[skiprowfoldF2])

data_target_training_subset2 = data_target.drop(data_target.index[skiprowfoldF2])

data_input_testing_subset2 = data_input[57:114]

data_target_testing_subset2 = data_target[57:114]


# fold 3
skiprowfoldF3 = [n for n in range(114,171)]
data_input_training_subset3 = data_input.drop(data_input.index[skiprowfoldF3])

data_target_training_subset3 = data_target.drop(data_target.index[skiprowfoldF3])

data_input_testing_subset3 = data_input[114:171]

data_target_testing_subset3 = data_target[114:171]


# fold 4
skiprowfoldF4 = [n for n in range(171,228)]
data_input_training_subset4 = data_input.drop(data_input.index[skiprowfoldF4])

data_target_training_subset4 = data_target.drop(data_target.index[skiprowfoldF4])

data_input_testing_subset4 = data_input[171:228]

data_target_testing_subset4 = data_target[171:228]


# fold 5
skiprowfoldF5 = [n for n in range(228,285)]
data_input_training_subset5 = data_input.drop(data_input.index[skiprowfoldF5])

data_target_training_subset5 = data_target.drop(data_target.index[skiprowfoldF5])

data_input_testing_subset5 = data_input[228:285]

data_target_testing_subset5 = data_target[228:285]


# fold 6
skiprowfoldF6 = [n for n in range(285,342)]
data_input_training_subset6 = data_input.drop(data_input.index[skiprowfoldF6])

data_target_training_subset6 = data_target.drop(data_target.index[skiprowfoldF6])

data_input_testing_subset6 = data_input[285:342]

data_target_testing_subset6 = data_target[285:342]


# fold 7
skiprowfoldF7 = [n for n in range(342,399)]
data_input_training_subset7 = data_input.drop(data_input.index[skiprowfoldF7])

data_target_training_subset7 = data_target.drop(data_target.index[skiprowfoldF7])

data_input_testing_subset7 = data_input[342:399]

data_target_testing_subset7 = data_target[342:399]


# fold 8
skiprowfoldF8 = [n for n in range(399,456)]
data_input_training_subset8 = data_input.drop(data_input.index[skiprowfoldF8])

data_target_training_subset8 = data_target.drop(data_target.index[skiprowfoldF8])

data_input_testing_subset8 = data_input[399:456]

data_target_testing_subset8 = data_target[399:456]


# fold 9
skiprowfoldF9 = [n for n in range(456,513)]
data_input_training_subset9 = data_input.drop(data_input.index[skiprowfoldF9])

data_target_training_subset9 = data_target.drop(data_target.index[skiprowfoldF9])

data_input_testing_subset9 = data_input[456:513]

data_target_testing_subset9 = data_target[456:513]


# fold 10
skiprowfoldF10 = [n for n in range(513,569)]
data_input_training_subset10 = data_input.drop(data_input.index[skiprowfoldF10])

data_target_training_subset10 = data_target.drop(data_target.index[skiprowfoldF10])

data_input_testing_subset10 = data_input[513:569]

data_target_testing_subset10 = data_target[513:569]
```

```python
def cross_validate_model(model):
    accuracies = []
    
    for i in range(1, num_subsets + 1):
        # Access variables directly using their names
        input_train_subset = globals()['data_input_training_subset{}'.format(i)]
        target_train_subset = globals()['data_target_training_subset{}'.format(i)]
        input_test_subset = globals()['data_input_testing_subset{}'.format(i)]
        target_test_subset = globals()['data_target_testing_subset{}'.format(i)]

        # Train the model
        model.fit(input_train_subset, target_train_subset)

        # Evaluate the model on the testing set and calculate accuracy
        test_predictions = model.predict(input_test_subset)
        test_accuracy = accuracy_score(test_predictions, target_test_subset)
        accuracies.append(test_accuracy)

    return accuracies
```

```python
def cross_validate_model_knn(model):
    accuracies = []
    
    for i in range(1, num_subsets + 1):
        # Access variables directly using their names
        input_train_subset = globals()['data_input_training_subset{}'.format(i)]
        target_train_subset = globals()['data_target_training_subset{}'.format(i)]
        input_test_subset = globals()['data_input_testing_subset{}'.format(i)]
        target_test_subset = globals()['data_target_testing_subset{}'.format(i)]

        # Train the model
        model.fit(input_train_subset, target_train_subset)

        # Evaluate the model on the testing set and calculate accuracy
        test_predictions = model.predict(input_test_subset.to_numpy())
        test_accuracy = accuracy_score(test_predictions, target_test_subset)
        accuracies.append(test_accuracy)

    return accuracies
```

```python
def model_svm():
    # Example usage with SVC
    # linear
    svc_model_linear = SVC(kernel='linear')
    svc_accuracies_linear = cross_validate_model(svc_model_linear)

    # rbf
    svc_model_rbf = SVC(kernel='rbf')
    svc_accuracies_rbf = cross_validate_model(svc_model_rbf)

    # poly
    svc_model_poly = SVC(kernel='poly')
    svc_accuracies_poly = cross_validate_model(svc_model_poly)

    # sigmoid
    svc_model_sigmoid= SVC(kernel='sigmoid')
    svc_accuracies_sigmoid = cross_validate_model(svc_model_sigmoid)

    # Print mean testing accuracies for each classifier
    print("SVC Linear Mean Testing Accuracy:", sum(svc_accuracies_linear) / num_subsets)
    print("SVC RBF Testing Accuracy:", sum(svc_accuracies_rbf) / num_subsets)
    print("SVC Poly Testing Accuracy:", sum(svc_accuracies_poly) / num_subsets)
    print("SVC Sigmoid Testing Accuracy:", sum(svc_accuracies_sigmoid) / num_subsets)
    
```

```python
def model_rf():
    # Example usage with RandomForestClassifier
    rf_model_100 = RandomForestClassifier(n_estimators=100, random_state=42)
    rf_accuracies_100 = cross_validate_model(rf_model_100)

    rf_model_200 = RandomForestClassifier(n_estimators=200, random_state=42)
    rf_accuracies_200 = cross_validate_model(rf_model_200)

    rf_model_500 = RandomForestClassifier(n_estimators=500, random_state=42)
    rf_accuracies_500 = cross_validate_model(rf_model_500)

    rf_model_1000 = RandomForestClassifier(n_estimators=1000, random_state=42)
    rf_accuracies_1000 = cross_validate_model(rf_model_1000)

    # Print mean testing accuracies for each classifier
    print("Random Forest (n_estimators=100) Mean Testing Accuracy:", sum(rf_accuracies_100) / num_subsets)
    print("Random Forest (n_estimators=200) Mean Testing Accuracy:", sum(rf_accuracies_200) / num_subsets)
    print("Random Forest (n_estimators=500) Mean Testing Accuracy:", sum(rf_accuracies_500) / num_subsets)
    print("Random Forest (n_estimators=1000) Mean Testing Accuracy:", sum(rf_accuracies_1000) / num_subsets)
```

```python
def model_knn():
    # Example usage with KNeighborsClassifier
    # n_neighbors=1
    knn1_model = KNeighborsClassifier(n_neighbors=1)
    knn1_accuracies = cross_validate_model_knn(knn1_model)

    # n_neighbors=5
    knn5_model = KNeighborsClassifier(n_neighbors=5)
    knn5_accuracies = cross_validate_model_knn(knn5_model)

    # n_neighbors=10
    knn10_model = KNeighborsClassifier(n_neighbors=10)
    knn10_accuracies = cross_validate_model_knn(knn10_model)

    # n_neighbors=15
    knn15_model = KNeighborsClassifier(n_neighbors=15)
    knn15_accuracies = cross_validate_model_knn(knn15_model)

    # Print mean testing accuracies for each classifier
    print("KNN 1 Mean Testing Accuracy:", sum(knn1_accuracies) / num_subsets)
    print("KNN 5 Mean Testing Accuracy:", sum(knn5_accuracies) / num_subsets)
    print("KNN 10 Mean Testing Accuracy:", sum(knn10_accuracies) / num_subsets)
    print("KNN 15 Mean Testing Accuracy:", sum(knn15_accuracies) / num_subsets)
```

```python
model_svm()

model_rf()

model_knn()
```
